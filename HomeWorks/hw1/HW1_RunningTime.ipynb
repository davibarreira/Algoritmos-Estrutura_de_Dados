```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import ast
```

```python
with open('sort_python.txt', 'r') as f:
    l = f.readlines()
l = ast.literal_eval(l[0])
df= pd.DataFrame(l)
```

```python
cpp_random = pd.read_csv('sort_cpp_random.txt',header=None)
cpp_random.columns = ['n','t1','t2','t3']

cpp_ascend = pd.read_csv('sort_cpp_ascend.txt',header=None)
cpp_ascend.columns = ['n','t1','t2','t3']

cpp_descend = pd.read_csv('sort_cpp_descend.txt',header=None)
cpp_descend.columns = ['n','t1','t2','t3']
```

```python
nsize = 10000
n     = np.arange(nsize,nsize*10+1,nsize)
col   = ['t1','t2','t3']
algo_cpp  = ['insert_sort','merge_sort', 'quick_sort']
fig = plt.figure(figsize=(12,5))
for i in range(3):
    ax1 = plt.subplot(121)
    ax1.plot(n,cpp_random[col[i]],label = algo_cpp[i]+' CPP')
    plt.ylabel('Tempo (s)')
    plt.xlabel('Tamanho do Vetor')
    
    ax2 = plt.subplot(122)
    ax2.plot(n,cpp_random[col[i]],label = algo_cpp[i]+' CPP')
    ax2.set_yscale('log')
    plt.ylabel('Tempo (s)')
    plt.xlabel('Tamanho do Vetor')
    
ax1.legend(bbox_to_anchor=(0,1.02),loc='lower left',ncol=3)
fig.savefig('SortTime_Random.png')
```
